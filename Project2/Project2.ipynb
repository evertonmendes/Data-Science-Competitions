```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df_train_transaction = pd.read_csv(PROJECT_PATH+'/Data/train_transaction.csv')
```

```python
numbers_0_1=df_train_transaction['isFraud']
```

```python
plot_label=""
for index in range(len(numbers_0_1.index)):
  plot_label+="{} : {} \n".format(numbers_0_1.index[index], round(numbers_0_1[index]/sum(numbers_0_1), 8))
```

```python
plt.title("Histogram of Fraud")
sn.histplot(df_train_transaction['isFraud'], label=plot_label)
plt.legend()
plt.savefig("Histogram.png")
```
